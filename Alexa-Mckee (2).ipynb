## <font color='pink'> The Analysis:
First, lets import our database...

```python
# Alexa Mckee
# CMPINF 0010 Spring 2023

import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

covid = pd.read_csv("covid-database.csv") #read in database
        
covid.head(100) #display first 100
```

> Thats a lot of information! Let's just look at the neighborhoods in Pittsburgh.. I also don't need all these columns, lets clean things up to only show top neighborhood infections

```python
covid_mask = covid['neighborhood_municipality'].str.contains("Pittsburgh") #filter mask to only show cases in pittsburgh

#deaths = covid[covid_mask] #apply mask
infections = covid[covid_mask]

columns = ["neighborhood_municipality","infections"] #only want these columns

for i in infections.columns:
    if i not in columns:
        del infections[i] #delete unneccessary columns from view

#covid.head(100)
sort = infections.sort_values( by="infections", ascending=False) #Sorting by most infections to date
print(sort)
```

```python
covidDict = {}

for index, row in infections.iterrows():

    name = row[1]
    quantity = row[0]

    if name in covidDict:
        covidDict[name] = covidDict[name] + quantity 
    else:
        covidDict[name] = quantity

for item, freq in covidDict.items():
    print(item,freq)
```

## <font color='pink'> Graphs

```python
# Plot dictionary
def plotDict(dict_name):
    #myList = dict_name.items()
    top10 = dict(list(sort[0: 10]))
    print(top10)
    #new = pd.DataFrame.from_dict(myList)
    #myList1 = myList.iloc[:90]
    
    x, y = zip(*top10)
    plt.plot(x, y)
    plt.xlabel('Neighborhood')
    plt.ylabel('Infections')
    plt.title('Top COVID-19 Infections by Pittsburgh Neightborhood')
    plt.xticks(rotation=90)
    plt.show()
    
plotDict(covidDict)
```
