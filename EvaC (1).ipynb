## Introduction:
We will be taking a look at the Pittsburgh Police Arrest Data and counting the number of arrests per neighborhood
## The Metric:
Our metric is the safest neighborhood, so the lowest amount of arrests will be awarded the safest neighborhood in this category.
Here is the first 10 arrests and their information below:

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

arrest = pd.read_csv("arrest.csv")
arrest.head(10)
```

As we said earlier, we are looking at which neighborhood has the fewest arrests, so we will be looking under the INCIDENTNEIGHBORHOOD column and counting each instance of a neighborhood up to identify the neighborhood with the fewest arrest numbers.

```python
a = arrest['INCIDENTNEIGHBORHOOD'].value_counts()
print(a)
```

Now lets make some graphs!

```python
arr = arrest['INCIDENTNEIGHBORHOOD']
arr = arrest['INCIDENTNEIGHBORHOOD'].value_counts().rename_axis("neighborhood").to_frame(name='arrests')
print(arr)
```

Now lets take a look at a graph of the bottom few neighborhoods when it comes to arrests, to determine which has the least amount of arrests.

```python
bottom_graph = arr.iloc[90:]


bottom_graph.plot.bar(title='Arrests in Neighborhoods in Pittsburgh')
plt.xlabel("Neighborhood")
plt.ylabel("Arrests")
```

## The Best Neighborhood!!
Based on this information alone, Mt. Oliver Neighborhood is the safest neighborhood.
